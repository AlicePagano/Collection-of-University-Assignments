```r
# Create list holding lake's name and all the parameters

lakes_list = list(c("Loch Ness", "Loch Lomond", "Loch Morar", "Loch Tay", 
               "Loch Awe", "Loch Maree", "Loch Ericht", 
               "Loch Lochy", "Loch Rannoch", "Loch Shiel", 
               "Loch Katrine", "Loch Arkaig", "Loch Shin"),
              c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35),
              c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5),
              c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8),
              c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49),
              c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)
             )
names(lakes_list) <- c("Loch", "Volume $[km^3]$", "Area $[km^2]$", "Length $[km]$", "Max. depth $[m]$", "Mean depth $[m]$")

# Create dataframe 
scottish.lakes <- data.frame(lakes_list)
# check attributes of the dataframe
attributes(scottish.lakes)

# Print dataframe
scottish.lakes
```

```r
# Evaluate the highest and lowest volume 
max_vol <- max(scottish.lakes$Volume)
min_vol <- min(scottish.lakes$Volume)

# Evaluate the highest and lowest area 
max_area <- max(scottish.lakes$Area)
min_area <- min(scottish.lakes$Area)

cat('Highest volume [km^3]:', max_vol, '\tLowest volume [km^3]:', min_vol, '\n')
cat('Highest area [km^2]:', max_area, '\tLowest area [km^2]:', min_area)
```

### 3.3 - Top 10 countries <a class="anchor" id="3.3"></a>
* [Index](#0)

```r
library(dplyr)
library(ggplot2)
```

```r
covid <- covid[order(covid$dateRep, decreasing=FALSE),]
```

```r
covid <- covid %>%
  group_by(countriesAndTerritories) %>%
  mutate("cm_death" = cumsum(deaths)) 
  

covid <- covid %>%
  group_by(countriesAndTerritories) %>%
  mutate("cm_cases" = cumsum(cases))

#mutate(group_by(df,id), cumsum=cumsum(value))
```

```r
#prova <- covid[covid$dateRep==date,]
#prova2 <- prova[prova$countriesAndTerritories=="Sweden",]
```

```r
today.covid <- covid[covid$dateRep==date,]

top10.today.covid <-today.covid[order(today.covid$cm_cases, decreasing=TRUE),][2:7,]

top10.today.covid 
```

```r
top10.covid <- covid[covid$countriesAndTerritories %in%
      top10.today.covid$countriesAndTerritories,]
```

```r
# Normalize date 
top10.covid <- top10.covid[top10.covid$cm_cases>0,]

#top10.covid[!duplicated(top10.covid$countriesAndTerritories),]

top10.covid <- top10.covid %>%
  group_by(countriesAndTerritories) %>%
  mutate(dateRepNorm = as.numeric(difftime(dateRep,first(dateRep), units="weeks")))     
```

```r
# Plot total number of cases as a function of time
pp1 <- ggplot(data=top10.covid, 
       aes(x=top10.covid$dateRepNorm, y=top10.covid$cm_cases, 
           color=top10.covid$countriesAndTerritories)) + geom_point(shape = 18) + labs(x = "Weeks", y="Number of cases", 
                                                                             title = "Total number of cases as a function of time",
                                                                             color = "Country") + theme_bw() + 
                                                                             scale_x_continuous(breaks = round(seq(min(top10.covid$dateRepNorm), 
                                                                                                max(top10.covid$dateRepNorm), by = 10),1))

# Plot total number of death as a function of time
pp2 <- ggplot(data=top10.covid, 
       aes(x=top10.covid$dateRepNorm, y=top10.covid$cm_death, 
           color=top10.covid$countriesAndTerritories)) + geom_point(shape = 18) + labs(x = "Weeks", y="Number of deaths", 
                                                                             title = "Total number of deaths as a function of time",
                                                                             color = "Country") + theme_bw() + 
                                                                             scale_x_continuous(breaks = round(seq(min(top10.covid$dateRepNorm), 
                                                                                                max(top10.covid$dateRepNorm), by = 10),1))

pp1
pp2
```
