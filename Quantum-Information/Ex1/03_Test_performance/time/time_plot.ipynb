```python
# Import main packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.palettes as palette # For palette of colors
import os
from mpl_toolkits import mplot3d

from sklearn.linear_model import LinearRegression

import bokeh.palettes as palette
```

## Load of the data 

```python
path = "files/time_cpu_1.txt"
data_cpu_1 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_2.txt"
data_cpu_2 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_3.txt"
data_cpu_3 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_matmul.txt"
data_cpu_matmul = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_1_O1.txt"
data_cpu_1_O1 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_1_O2.txt"
data_cpu_1_O2 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_1_O3.txt"
data_cpu_1_O3 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_1_Ofast.txt"
data_cpu_1_Ofast = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_2_O1.txt"
data_cpu_2_O1 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_2_O2.txt"
data_cpu_2_O2 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_2_O3.txt"
data_cpu_2_O3 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_2_Ofast.txt"
data_cpu_2_Ofast = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_3_O1.txt"
data_cpu_3_O1 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_3_O2.txt"
data_cpu_3_O2 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_3_O3.txt"
data_cpu_3_O3 = pd.read_csv(path,sep="\s+",skiprows=0)

path = "files/time_cpu_3_Ofast.txt"
data_cpu_3_Ofast = pd.read_csv(path,sep="\s+",skiprows=0)
```

## Plots 

```python
# Plot of experimental results ##DA INSERIRE E INSERIRE TARGHE INDICI
fig, (ax1) = plt.subplots(1, 1, figsize=(10,8))

col = palette.inferno(5)

#ax1.plot(data['n'],data['time'],'.',color=col[1])
ax1.plot(data_cpu_3['n'], data_cpu_3['time(s)'], color=col[0],label='Order 3')
ax1.plot(data_cpu_1['n'], data_cpu_1['time(s)'], color=col[1],label='Order 1') 
ax1.plot(data_cpu_2['n'], data_cpu_2['time(s)'], color=col[2],label='Order 2')
ax1.plot(data_cpu_matmul['n'], data_cpu_matmul['time(s)'], color=col[3],label='Matmul')

#ax1.set_title(r'Plot of absorbtion as a function of the wavelength',fontsize=20)
ax1.set_xlabel(r'Matrix size',fontsize=18)
ax1.set_ylabel(r'Computational time [sec]',fontsize=18)
ax1.grid(color='0.93')
ax1.legend(title='',fontsize=18,  title_fontsize=18,loc='best')
ax1.set_title('Matrix multiplication',fontsize=20)
ax1.tick_params(axis="x", labelsize=16)
ax1.tick_params(axis="y", labelsize=16)
#ax1.set_xscale('log')
ax1.set_yscale('log')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.tight_layout()

if not os.path.isdir('./plots'):
    os.mkdir('./plots')

plt.savefig('plots/time.pdf', format='pdf')
#plt.close()
```
